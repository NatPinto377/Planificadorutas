```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear el grafo y definir conexiones entre estaciones con su peso correspondiente
def crear_grafo():
    G = nx.Graph()

    # Conexiones entre estaciones (con pesos como distancia o tiempo de viaje)
    G.add_edge("A", "B", peso=2)  # Conexión entre A y B con un peso de 2
    G.add_edge("B", "C", peso=3)  # Conexión entre B y C con un peso de 3
    G.add_edge("C", "D", peso=2)  # Conexión entre C y D con un peso de 2
    G.add_edge("A", "D", peso=5)  # Conexión directa entre A y D con un peso de 5
    G.add_edge("B", "E", peso=4)  # Conexión entre B y E con un peso de 4
    G.add_edge("E", "D", peso=1)  # Conexión entre E y D con un peso de 1

    return G

# Clase PlanificadorRutas con el algoritmo de Dijkstra
class PlanificadorRutas:
    def __init__(self, grafo):
        self.grafo = grafo

    def encontrar_mejor_ruta(self, origen, destino):
        """Encuentra la mejor ruta usando el algoritmo de Dijkstra"""
        try:
            # Usar el algoritmo de Dijkstra para encontrar el camino más corto
            ruta = nx.dijkstra_path(self.grafo, origen, destino, weight='peso')
            return ruta
        except nx.NetworkXNoPath:
            # Si no hay ruta entre origen y destino
            return None

# Función para representar el grafo de manera gráfica
def visualizar_grafo(grafo):
    pos = nx.spring_layout(grafo)  # Organiza los nodos de forma distribuida
    plt.figure(figsize=(8, 6))  # Tamaño de la figura

    # Dibuja los nodos y las aristas
    nx.draw(grafo, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=12, font_weight="bold", edge_color="gray")

    # Dibuja los pesos de las aristas
    edge_labels = nx.get_edge_attributes(grafo, 'peso')
    nx.draw_networkx_edge_labels(grafo, pos, edge_labels=edge_labels, font_size=12)

    # Título y mostrar la figura
    plt.title("Representación Gráfica del Grafo de Estaciones")
    plt.show()

# Función principal
def main():
    # Crear el grafo
    G = crear_grafo()

    # Inicializar el planificador de rutas con el grafo creado
    planificador = PlanificadorRutas(G)

    # Definir los puntos de prueba (origen, destino)
    puntos_prueba = [("A", "D"), ("B", "E"), ("A", "C"), ("E", "D"), ("C", "A")]

    # Ejecutar las pruebas y mostrar las mejores rutas
    for origen, destino in puntos_prueba:
        ruta = planificador.encontrar_mejor_ruta(origen, destino)
        if ruta:
            print(f"Mejor ruta encontrada de {origen} a {destino}: {' -> '.join(ruta)}")
        else:
            print(f"No existe una ruta entre {origen} y {destino}.")

    # Visualizar el grafo
    visualizar_grafo(G)

# Ejecutar el programa principal
if __name__ == "__main__":
    main()
```
